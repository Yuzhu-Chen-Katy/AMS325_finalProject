```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from pandas.plotting import scatter_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read the data
data_1 = pd.read_csv('C:/Users/katyc/Downloads/data.csv')
data_2 = pd.read_csv('C:/Users/katyc/Downloads/data_2genre.csv')
#combine two datasets
data = pd.concat([data_1, data_2])
```

```python
data['label'].value_counts()
```

```python

def plot_genre_tempo(data, tempo_col='tempo', genres=['jazz', 'pop', 'classical', 'hiphop', 'disco', 'country', 'rock', 'metal', 'reggae', 'blues'], figsize=(30, 10)):
    plt.figure(figsize=figsize)

    for genre in genres:
        sns.kdeplot(data=data.loc[data['label'] == genre, tempo_col], label=genre)

    plt.title("Distribution of tempos by genre", fontsize=18)
    plt.xlabel("Tempo", fontsize=18)
    plt.legend()

# Example usage with a custom tempo column 'custom_tempo'
plot_genre_tempo(data, tempo_col='tempo')
plt.show()
```

```python
plt.figure(figsize=(30,10))

genres = data['label'].unique()[:-2]

tempos = [ data[data['label']==x].tempo.mean() for x in genres ]

sns.barplot(x=genres, y=tempos, palette="deep")

plt.title("Average tempo by genre", fontsize = 18)

plt.xlabel('Genre', fontsize = 18)
plt.ylabel('Mean Tempo', fontsize = 18)
```

```python

def plot_genre_spect(data, tempo_col='spectral', genres=['jazz', 'pop', 'classical', 'hiphop', 'disco', 'country', 'rock', 'metal', 'reggae', 'blues'], figsize=(30, 10)):
    plt.figure(figsize=figsize)

    for genre in genres:
        sns.kdeplot(data=data.loc[data['label'] == genre, tempo_col], label=genre)

    plt.title("Distribution of spectral centroids by genre", fontsize=18)
    plt.xlabel("Spectral Centroids", fontsize=18)
    plt.legend()

# Example usage with a custom tempo column 'custom_tempo'
plot_genre_spect(data, tempo_col='spectral_centroid')
plt.show()
```
